Realizo un grafico de suma de ventas diarias en ambos meses para poder analizarlo.


```python
import pandas as pd
import matplotlib.pyplot as plt

df_abril_mayo =df[(df['Venta Fecha Carga'].dt.month==4) | (df['Venta Fecha Carga'].dt.month==5)]
ventas_por_mes =df_abril_mayo.groupby('Venta Fecha Carga')['Venta ID'].count()

Q1 =ventas_por_mes.quantile(0.25)
Q3 =ventas_por_mes.quantile(0.75)
IQR =Q3 - Q1
lower_bound =Q1 - 1.5 * IQR
upper_bound =Q3 + 1.5 * IQR
ventas_por_mes =ventas_por_mes[(ventas_por_mes >= lower_bound) & (ventas_por_mes <= upper_bound)]

plt.figure(figsize=(12, 6))
plt.scatter(ventas_por_mes.index, ventas_por_mes.values)


plt.xlabel('Venta Fecha Carga')
plt.ylabel('Venta ID')

plt.title('Ventas de Abril y Mayo sin outliers')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans =KMeans(n_clusters=3)

kmeans.fit(ventas_por_mes.values.reshape(-1, 1))
clusters =kmeans.predict(ventas_por_mes.values.reshape(-1, 1))

plt.figure(figsize=(12, 6))
plt.scatter(ventas_por_mes.index, ventas_por_mes.values, c=clusters)

plt.xlabel('Venta Fecha Carga')
plt.ylabel('Ventas')

plt.title('Ventas de Abril y Mayo con clusters')
plt.show()

print("El primer cluster contiene los siguientes datos:")
print(ventas_por_mes[clusters == 0])

print("El segundo cluster contiene los siguientes datos:")
print(ventas_por_mes[clusters == 1])

print("El tercer cluster contiene los siguientes datos:")
print(ventas_por_mes[clusters == 2])
```

Creo clusters en cuanto a la cantidad de ventas , usando lo aprendido sobre sklearn y probando su funcionalidad.

```python
df_junio =df[df['Venta Fecha Carga'].dt.month==6]

ventas_junio = df_junio.groupby('Venta Fecha Carga')['Venta ID'].count()
cluster = kmeans.predict(ventas_junio.values.reshape(-1, 1))
print(f"El mes 6 pertenece al cluster {cluster[0]}")

plt.figure(figsize=(12, 6))
plt.scatter(ventas_por_mes.index, ventas_por_mes.values, c=clusters)
plt.scatter(ventas_junio.index, ventas_junio.values, c='red')

plt.xlabel('Venta Fecha Carga')
plt.ylabel('Ventas')
plt.title('Ventas de Abril, Mayo y Junio con clusters')
plt.show()
```

Agrupo los clusters


```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

df_abril_mayo =pd.read_excel('/content/drive/MyDrive/Data Science Proyecto Final/DATASET.xlsx')
df_abril_mayo =df_abril_mayo[(df_abril_mayo['Venta Fecha Carga'].dt.month==4) | (df_abril_mayo['Venta Fecha Carga'].dt.month==5)]

features = ['Venta ID', 'Precio Venta']

scaler =MinMaxScaler()
scaled_data =scaler.fit_transform(df_abril_mayo[features])
kmeans =KMeans(n_clusters=3)
kmeans.fit(scaled_data)
clusters =kmeans.predict(scaled_data)

#DATA FRAME NUEVO
df_junio =pd.read_excel('/content/drive/MyDrive/Data Science Proyecto Final/DATASET.xlsx')
df_junio =df_junio[df_junio['Venta Fecha Carga'].dt.month==6]

features =['Venta ID', 'Precio Venta']
scaled_data_junio =scaler.transform(df_junio[features])
cluster =kmeans.predict(scaled_data_junio)
print(f"El mes 6 pertenece al cluster {cluster[0]}")

plt.figure(figsize=(12, 6))
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=clusters)
plt.scatter(scaled_data_junio[:, 0], scaled_data_junio[:, 1], c='red')
plt.xlabel('Ventas')
plt.ylabel('Precio Venta')
plt.title('Ventas de Abril, Mayo y Junio con clusters')
plt.show()
```
